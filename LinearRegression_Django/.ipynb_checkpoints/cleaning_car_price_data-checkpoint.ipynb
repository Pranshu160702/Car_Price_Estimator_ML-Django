# Issues Related To Data

```python
# As you can see year has some unwanted values like SALE, cen, sumo, etc.
# price also has some unwanted values like 'Ask for Price'
# fuel type and kms driven has some nan values
# name is too long so lets just keep the first 3 words of it
# Also year and price are object type but they should be int type
# Lastly kms_driven unwanted 'kms' attached to it and commas between int values
```

```python
new_car_data = car_data.copy()
```

# Cleaning Data

```python
# Removing non numerical values
new_car_data = car_data[car_data['year'].str.isnumeric()] # Returns true if numeric or else false hence gives rows that are numeric only
```

```python
# Converting object type to integer
new_car_data['year'] = new_car_data['year'].astype(int)
```

```python
new_car_data['year'].unique()
```

```python
# Removing non numerical values in price
new_car_data = new_car_data[new_car_data['Price']!="Ask For Price"]
```

```python
new_car_data
```

```python
# Removing the commas in price
new_car_data['Price'] = new_car_data['Price'].str.replace(',','').astype(int)
```

```python
new_car_data
```

```python
new_car_data.info()
```

```python
# Remobing the kms after values in kms_driven
new_car_data['kms_driven'] = new_car_data['kms_driven'].str.split(' ').str.get(0)
```

```python
new_car_data['kms_driven']
```

```python
# Removing commas
new_car_data['kms_driven'] = new_car_data['kms_driven'].str.replace(',','')
```

```python
new_car_data['kms_driven']
```

```python
# Removing non numeric values in kms_driven
new_car_data = new_car_data[new_car_data['kms_driven'].str.isnumeric()]
```

```python
# Converting object type to int type
new_car_data['kms_driven'] = new_car_data['kms_driven'].astype(int)
```

```python
new_car_data
```

```python
# Removing NaN values from fuel type ( ~ means not )
new_car_data = new_car_data[~new_car_data['fuel_type'].isna()]
```

```python
# Keeping only 3 words for name of car
new_car_data['name'] = new_car_data['name'].str.split(' ').str.slice(0,3).str.join(' ')
```

```python
new_car_data
```

```python
# Removing Outliers
new_car_data = new_car_data[new_car_data['Price'] < 6e6] # Price less than 6 lakhs
```

```python
# Resetting the final indexs are dropping old ones
new_car_data =  new_car_data.reset_index(drop=True)
```

```python
new_car_data
```

```python
# Saving this data as a new csv
new_car_data.to_csv('cleaned_car_data.csv', index=False)
```
